```python
import pandas as pd 
```

```python
df = pd.read_csv('/workspaces/docker-homework/taxi_zone_lookup.csv')
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://postgres:postgres@localhost:5433/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='zones', con=engine))
```

```python
# Create table schema (no data)
df.head(0).to_sql(
    name="zones",
    con=engine,
    if_exists="replace"
)
first = False
print("Table created")

# Insert chunk
df.to_sql(
    name="zones",
    con=engine,
    if_exists="append"
)

print("Inserted:", len(df))
```
